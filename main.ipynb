`````python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import os
import gzip
`````

Definición de las categorías

`````python
categories = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
              'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
`````

Función para cargar los datos de MNIST Fashion desde los archivos .gz.

Las imagenes y su respectiva etiqueta estan separados en archivos .gz para entrenamiento y prueba.

`````python
def load_mnist(path, kind='train'):

    """Load MNIST data from `path`"""
    labels_path = os.path.join(path, f"{kind}-labels-idx1-ubyte.gz")
    images_path = os.path.join(path, f"{kind}-images-idx3-ubyte.gz")

    with gzip.open(labels_path, 'rb') as lbpath:
        labels = np.frombuffer(lbpath.read(), dtype=np.uint8,
                               offset=8)

    with gzip.open(images_path, 'rb') as imgpath:
        images = np.frombuffer(imgpath.read(), dtype=np.uint8,
                               offset=16).reshape(len(labels), 28, 28)  # Reshape to 28x28

    return images, labels
`````

Cargamos los datos de entrenamiento y prueba del respectivo archivo .gz

`````python
X_train, y_train = load_mnist('data/fashion', kind='train')
X_test, y_test = load_mnist('data/fashion', kind='t10k')
`````

Normalizar los pixeles de las imagenes para que esten en el rango [0, 1] en lugar de [0, 255]

`````python
def normalize(img, tag):
    img = tf.cast(img, tf.float32)
    return img / 255.0, tag


train_dataset = tf.data.Dataset.from_tensor_slices((X_train, y_train)) # Crea un dataset a partir de los datos
train_dataset = train_dataset.map(normalize) # Normaliza los datos

test_dataset = tf.data.Dataset.from_tensor_slices((X_test, y_test)) # Crea un dataset a partir de los datos
test_dataset = test_dataset.map(normalize) # Normaliza los datos
`````

Visualización de las primeras 25 imagenes del dataset de entrenamiento con su respectiva etiqueta

`````python
plt.figure(figsize=(10, 10))
i = 0
for image, label in train_dataset.take(25):
    image = image.numpy().reshape(28, 28)
    plt.subplot(5, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(image, cmap="binary")
    plt.xlabel(categories[label])
    i += 1
plt.show()
`````

### Creación del modelo secuencial
1. Capa de entrada: Los datos de entrada son imagenes de 28x28.
2. Capa de aplanado: Los datos de entrada son aplanados de 2D (28x28) a 1D (784 neuronas).
3. Capa densa: 50 neuronas con función de activación ReLU.
4. Capa densa: 50 neuronas con función de activación ReLU.
5. Capa de salida: 10 neuronas con función de activación Softmax una para cada categoría.

Por lo tanto, el modelo tiene una capa de entrada de 784 neuronas, dos capas ocultas de 50 neuronas y una capa de salida de 10 neuronas.

`````python
model = tf.keras.Sequential([
    tf.keras.Input(shape=(28, 28)), # Input: 28x28
    tf.keras.layers.Flatten(),  # Flatten: 2D -> 1D
    tf.keras.layers.Dense(50, activation=tf.nn.relu),
    tf.keras.layers.Dense(50, activation=tf.nn.relu),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)  # Softmax: Asegura que la suma de las posibilidades sea 1 o 100%
])
`````

### Compilación del modelo
1. Optimizador: Adam, es un optimizador que ajusta la tasa de aprendizaje de forma adaptativa.
2. Función de pérdida: sparse_categorical_crossentropy, es una función de pérdida que mide la diferencia entre las etiquetas reales y las predicciones.
3. Métricas: accuracy, mide la precisión del modelo.

`````python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
`````

### Preparación de los datos para el entrenamiento
1. Se define el tamaño del lote (batch) en 32. Es la cantidad de ejemplos que se utilizan en una iteración del entrenamiento antes de actualizar los pesos.
2. El conjunto de datos se repite indefinidamente para que el modelo vea todas las imagenes de entrenamiento.
3. Se mezclan 60000 imagenes (tamaño del buffer) para que el modelo no aprenda de la misma forma las imagenes.
4. Se agrupan las imagenes en lotes de 32.
5. El conjunto de datos de prueba se agrupa en lotes de 32.

`````python
LOT_SIZE = 32

train_dataset = train_dataset.repeat().shuffle(buffer_size=60000).batch(LOT_SIZE)
test_dataset = test_dataset.batch(LOT_SIZE)
`````

### Entrenamiento del modelo
1. Se entrena el modelo con el conjunto de datos de entrenamiento.
2. Se realizan 5 iteraciones (epochs) sobre el conjunto de datos de entrenamiento, donde en cada iteración se mezclan las imagenes.
3. Se especifica el número de pasos por época (steps_per_epoch) que es el número de lotes totales en el conjunto de datos de entrenamiento para que el modelo vea todas las imagenes.

`````python
history = model.fit(train_dataset, epochs=5, steps_per_epoch=60000 // LOT_SIZE)
`````

### Pérdida del modelo
Se grafica la pérdida del modelo en cada época, entre más avanzan las épocas la pérdida disminuye.

`````python
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.plot(history.history['loss'])
`````

### Predicción del modelo
Se evalua el conjunto de datos de prueba con el modelo entrenado y se compara la predicción con la etiqueta real.

La cantidad de imagenes que se evaluan se define en ````num_images````.

23 imagenes son clasificadas correctamente y 2 incorrectamente, por lo que la precisión es bastante alta.

`````python
image = np.array([])
label = np.array([])
predictions = np.array([])

for test_img, test_tag in test_dataset.take(1):
    image = test_img.numpy()
    label = test_tag.numpy()
    predictions = model.predict(image)


# Función para graficar la imagen, la etiqueta correcta y la predicción
def plot_image(index, predictions_array, true_label, img):
    predictions_array, true_label, img = predictions_array[index], true_label[index], img[index]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])

    plt.imshow(img, cmap="binary")

    predicted_label = np.argmax(predictions_array)
    if predicted_label == true_label:
        color = 'blue'
    else:
        color = 'red'

    plt.xlabel("{} {:2.0f}% ({})".format(categories[predicted_label],
                                         100 * np.max(predictions_array),
                                         categories[true_label]),
               color=color)
    

# Función para graficar dentro de la predicción la probabilidad de cada categoría y la categoría con mayor probabilidad
def plot_value_array(index, predictions_array, true_label):
    predictions_array, true_label = predictions_array[index], true_label[index]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    thisplot = plt.bar(range(10), predictions_array, color="#777777")
    plt.ylim([0, 1])
    predicted_label = np.argmax(predictions_array)

    thisplot[predicted_label].set_color('red')
    thisplot[true_label].set_color('blue')
    

num_rows = 5
num_cols = 5
num_images = num_rows * num_cols
plt.figure(figsize=(2 * 2 * num_cols, 2 * num_rows))
for i in range(num_images):
    plt.subplot(num_rows, 2 * num_cols, 2 * i + 1)
    plot_image(i, predictions, label, image)
    plt.subplot(num_rows, 2 * num_cols, 2 * i + 2)
    plot_value_array(i, predictions, label)
`````
